```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
titles = []
expanded_times = []
# Make a request to the 
for i in range(1,20):
  url = f'https://www.livemint.com/market/stock-market-news/page-{i}'
  webpage = requests.get(url).text

# Parse the HTML content of the webpage using Beautiful Soup

  soup=BeautifulSoup(webpage,'lxml')
  soup.find_all('div',class_='listtostory clearfix')

  divs = soup.find_all('div', class_='listtostory')
  for div in divs:
      title = div.find('h2', class_='headline').text.strip()
      expanded_time = div.find('span', {'data-expandedtime': True})['data-expandedtime']
      titles.append(title)
      expanded_times.append(expanded_time)
df = pd.DataFrame()
df['news_titles'] = titles
df['time']= expanded_times
```
